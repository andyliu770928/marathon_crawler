## 1. 取出今天所有的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import datetime
# 打開瀏覽器
browser = webdriver.Chrome(executable_path='../chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")

# 每個兩秒鐘自動往下滑
import time
for i in range(2):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    
# 取得資料，丟到 BeautifulSoup 解析
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
print(datetime.datetime.now())
for d in soup.find(class_="part_list_2").find_all('h3'):
    if '2019/12/26' in d.find(class_="date").text:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)

# 關閉瀏覽器
browser.quit();
```

## 2. 取出現在時間兩小時內的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import datetime
# 打開瀏覽器
browser = webdriver.Chrome(executable_path='../chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")

# 每個兩秒鐘自動往下滑
import time
for i in range(2):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    
# 取得資料，丟到 BeautifulSoup 解析
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
end_time = datetime.datetime.now()-datetime.timedelta(hours=2)
end_time_str = datetime.datetime.strftime(end_time,'%Y/%m/%d %H:%M')
for d in soup.find(class_="part_list_2").find_all('h3'):
    if d.find(class_="date").text > end_time_str:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)

# 關閉瀏覽器
browser.quit();
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import datetime
# 打開瀏覽器
browser = webdriver.Chrome(executable_path='../chromedriver')
browser.get("https://www.ettoday.net/news/news-list-2019-12-23-0.htm")

# 每個兩秒鐘自動往下滑
import time
for i in range(40):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight),2000;")
    
    
# 取得資料，丟到 BeautifulSoup 解析
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
start_time_str = '2019/12/23 15:00'
end_time_str = '2019/12/23 17:00'
for d in soup.find(class_="part_list_2").find_all('h3'):
    if d.find(class_="date").text < start_time_str:
        break
    if d.find(class_="date").text <= end_time_str:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)

# 關閉瀏覽器
browser.quit();
```
