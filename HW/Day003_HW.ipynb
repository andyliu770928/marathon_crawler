### 比較一下範例檔案中的「File I/O」與「xmltodict」讀出來的內容有什麼差異


File I/O 讀出來的格式比較像是html，而xmltodict則是像是dict的格式

### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
import xmltodict
```

```python
# 1. 請問高雄市有多少地區有溫度資料？

fh = open("./data/64_72hr_CH.xml", "r")
xml = fh.read()
fh.close()

d = dict(xmltodict.parse(xml))

dataset = d['cwbopendata']['dataset']['locations']['location']

print(len(dataset))
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度

for i in range(len(dataset)):
    for key, val in dataset[i].items():
        if key == 'locationName':
            location = val
        elif key == 'weatherElement':
            print(location,val[0]['time'][0]['dataTime'],val[0]['time'][0]['elementValue']['value'])
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度

for key, val in dataset[0].items():
    if key == 'locationName':
        location = val
    elif key == 'weatherElement':
        for j in range(len(val)):
            print(location,val[0]['time'][j]['dataTime'],val[0]['time'][0]['elementValue']['value'])
```
